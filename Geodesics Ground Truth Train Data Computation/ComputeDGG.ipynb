```python
import numpy as np
import pickle
meshvertices = np.load("MeshSMALVerticesRandom.npz")['arr_0']
with open("smal_CVPR2017.pkl", "rb") as f:
    smal = pickle.load(f, encoding="latin1")
    faces = smal['f']
```

```python
#Compute Geodesics Dataset for Training Data using SMAL Model
import subprocess
from tqdm import tqdm
import os
import trimesh

for i in tqdm(range(0,meshvertices.shape[0])):
    #if(os.path.exists("DGGData" + os.sep + f"human_{i}_FD0.0100000000_c20.binary")):
    #    continue
    #if(i not in validx):
    #    continue
    mesh_i = meshvertices[i]
    mesh_i = trimesh.Trimesh(vertices = mesh_i, faces=faces)
    mesh_i.export(f"DGGData\\smal_{i}.obj")
    subprocess.call(f"DGG_precompute.exe DGGData\\smal_{i}.obj 0.01 f 20 2")
    os.remove(f"DGGData\\smal_{i}.obj")
    #break
```
